```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\murali\Downloads\customer_churn.csv")
```

```python
data.info()
```

check for imbalance

```python
data['Churn'].value_counts()
```

```python
data.drop(columns=['customerID'],axis=1,inplace=True)
```

```python
for col in data.columns:
    print(f"\nColumn: {col}")
    print(data[col].value_counts())
```

```python
data.replace('No internet service','No',inplace=True)
data.replace('No phone service','No',inplace=True)
```

```python
data.replace({'Yes':1,'No':0},inplace=True)
```

```python
data
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
data.info()
```

```python
data['gender'].replace({"Male":1,"Female":0},inplace=True)
```

```python
data=pd.get_dummies(data,columns=['InternetService','Contract','PaymentMethod'],drop_first=True)
```

```python
data
```

```python
data.info()
```

tenure,monthlycharges and totalcharges needs to be scaled

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

cols=['tenure','MonthlyCharges','TotalCharges']
data[cols] = scaler.fit_transform(data[cols])
```

```python
data
```

now imbalance needs to handled


```python
minority_class = data[data['Churn'] == 1]  # or == 'Yes' based on your encoding
minority=minority_class.sample(n=4000,replace=True)
```

```python
minority['Churn'].value_counts()
```

```python
majorrity_class=data[data['Churn']==0]
majority=majorrity_class.sample(n=4000)
```

```python
majority['Churn'].value_counts()
```

```python
final=pd.concat([majority,minority])
```

```python
final.reset_index(drop=True)
```

```python
X=final.drop(['Churn'],axis=1)
y=final['Churn']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential([
    Dense(100, input_shape=(23,), activation='relu'),
    Dropout(rate=0.1),
    Dense(100,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

just train the model for small epochs

```python
model.fit(X_train,y_train,epochs=7)
```

```python
model.evaluate(X_test,y_test)
```

model accuracy is just 48% because of the oversampling technique
